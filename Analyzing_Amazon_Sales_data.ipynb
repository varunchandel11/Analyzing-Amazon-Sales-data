```python
DaySalesInsights = df.copy()
DaySalesInsights['Invoice_Date'] = pd.to_datetime(df['Order Date']).dt.date
```

```python
# Grouping by 'Invoice_Date' and summing up the total revenue for each date
top10sales = DaySalesInsights.groupby('Invoice_Date')['Total Revenue'].sum()

# Sorting the grouped data in descending order based on the sum of total revenue
top10sales = top10sales.sort_values(ascending=False)

# Selecting the top 10 entries (i.e., the top 10 days with the highest total revenue)
top10sales = top10sales.head(10)

# Resetting the index to convert the 'Invoice_Date' from the index to a column
top10sales = top10sales.reset_index()

# Displaying the top 10 days with the highest total revenue
print(top10sales)
```

```python
sns.catplot(y = 'Total Revenue', x = 'Invoice_Date', data = top10sales, aspect = 2,palette='turbo',kind="bar")
plt.title('Top 10 Days When Sales Were Highest')
top10sales[['Total Revenue']]
plt.show()
```

```python
plt.figure(figsize=(18,7))
plt.pie('Total Revenue',labels='Invoice_Date',data = top10sales,
        autopct='%1.2f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.title('Contribution Of Sales Amount Among 10 Days')
plt.legend(round(top10sales['Total Revenue'],2), loc=7, fontsize = 'x-large')
plt.show()
```

```python
# Grouping by 'Invoice_Date' and summing up the total profits for each date
top10profits = DaySalesInsights.groupby('Invoice_Date')['Total Profit'].sum()

# Sorting the grouped data in descending order based on the sum of total profits
top10profits = top10profits.sort_values(ascending=False)

# Selecting the top 10 entries (i.e., the top 10 days with the highest total profits)
top10profits = top10profits.head(10)

# Resetting the index to convert the 'Invoice_Date' from the index to a column
top10profits = top10profits.reset_index()

# Displaying the top 10 days with the highest total profits
print(top10profits)
```

```python
sns.catplot(y = 'Total Profit', x = 'Invoice_Date', data = top10profits, aspect = 2,palette='turbo',kind="bar")
plt.title('Top 10 Days When Profits Were Highest')
top10profits[['Total Profit']]
```

```python
# Plotting
plt.figure(figsize=(18, 7))
plt.pie(top10profits['Total Profit'], labels=top10profits['Invoice_Date'], autopct='%1.2f%%', shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Contribution Of Profits Among Top 10 Days')
plt.legend(round(top10profits['Total Profit'], 2), loc='center left', fontsize='x-large', title='Total Profit')
plt.show()

```

```python
# Grouping by 'Invoice_Date' and summing up the units sold for each date
highqty = DaySalesInsights.groupby('Invoice_Date')['Units Sold'].sum()

# Sorting the grouped data in descending order based on the sum of units sold
highqty = highqty.sort_values(ascending=False)

# Selecting the top 10 entries (i.e., the top 10 days with the highest units sold)
highqty = highqty.head(10)

# Resetting the index to convert the 'Invoice_Date' from the index to a column
highqty = highqty.reset_index()

# Displaying the top 10 days with the highest units sold
print(highqty)
```

```python
# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='Invoice_Date', y='Units Sold', data=highqty, palette='turbo')
plt.title('Top 10 Days When Highest Quantity Of Items Were Sold')
plt.xlabel('Invoice Date')
plt.ylabel('Sales Quantity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Grouping by 'Order Date' and counting the number of orders for each date
MostOrders = DaySalesInsights.groupby('Order Date')['Order ID'].count()

# Sorting the grouped data in descending order based on the count of orders
MostOrders = MostOrders.sort_values(ascending=False)

# Selecting the top 10 entries (i.e., the top 10 days with the highest number of orders)
MostOrders = MostOrders.head(10)

# Resetting the index to convert the 'Order Date' from the index to a column
MostOrders = MostOrders.reset_index()

# Displaying the top 10 days with the highest number of orders
print(MostOrders)
```

```python
# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='Order Date', y='Order ID', data=MostOrders, palette='turbo')
plt.title('Top 10 Days When Most Orders Were Placed')
plt.xlabel('Order Date')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.pie(x='Order ID', labels='Order Date', data=MostOrders, autopct='%1.2f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Distribution of Orders Among Top 10 Days')
plt.legend(MostOrders['Order ID'], loc='center left', fontsize='x-large', title='Order ID')
plt.show()
```
